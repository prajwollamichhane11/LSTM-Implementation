```python
import tensorflow as tf
from tensorflow.keras.models import Sequential 
import keras
from keras.layers import Dense
from keras.layers import LSTM
from sklearn.model_selection import train_test_split
import numpy as np
import matplotlib.pyplot as plt
```

```python
#DATA PreProcessing
```

```python
def generateDATA():
    Data = []
    for j in range(100):
        eachInstance = []
        for i in range(5):
            eachInstance.append([(i+j)/100])
        Data.append(eachInstance)
    return Data

def generateTarget():
    target = []
    for i in range(100):
        target.append((i+5)/100)
    return target
```

```python
Data = generateDATA()
target = generateTarget()
```

```python
data = np.array(Data, dtype=float)
target = np.array(target, dtype=float)
```

```python
print(f"Data Shape->{data.shape}",f"Target Shape->{target.shape}")
```

```python
#Splitting the Data
x_train, x_test, y_train, y_test = train_test_split(data,target,test_size=0.2,random_state=4)
```

```python
#Implementing the model
# RNN Model
```

```python
model = keras.Sequential()
model.add(LSTM((1),batch_input_shape=(None,5,1),return_sequences=True)) #output_shape,batch,return_sequence(it gives output in each hidden state)
model.add(LSTM((1),return_sequences=False))
```

```python
model.compile(loss="mean_absolute_error",optimizer="adam",metrics=["accuracy"])
```

```python
model.summary()
```

```python
history = model.fit(x_train,y_train,epochs=500,validation_data=(x_test,y_test))
```

```python
results = model.predict(x_test)
```

```python
plt.scatter(range(20),results,c='r')
plt.scatter(range(20),y_test,c='g')
```

```python
plt.plot(history.history['loss'])
```

We can see the model is not converging and the loss is very high too. So, this migh be because we have not normalized the data. Lets, normalize the data by dividing by 100 on Data and target.
